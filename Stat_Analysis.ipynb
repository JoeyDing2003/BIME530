```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
over60 =pd.read_csv("/Users/Joey/Downloads/BIME530_Assignment1/Athena/over60.csv")
under60 =pd.read_csv("/Users/Joey/Downloads/BIME530_Assignment1/Athena/under60.csv")
```

## Calculate P-value

```python
# use scipy to calculate p_value for each arrary: under60["drug_exposure_cycle"] vs over60["drug_exposure_cycle"]
under60_array_drug_cycle = np.array(under60["drug_exposure_cycle"])
over60_array_drug_cycle = np.array(over60["drug_exposure_cycle"])
t_stat, p_value = stats.ttest_ind(under60_array_drug_cycle, over60_array_drug_cycle, equal_var=True)
print("P-value:", p_value)
```

```python
# use scipy to calculate p_value for each arrary: under60["average_days_supply"] vs over60["average_days_supply"]
under60_array_days_supply = np.array(under60["average_days_supply"])
over60_array_days_supply = np.array(over60["average_days_supply"])
t_stat, p_value = stats.ttest_ind(under60_array_days_supply, over60_array_days_supply, equal_var=True)
print("P-value:", p_value)
```

## Graphically summary

```python
# The distribution drug_usage
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Drug exposure cycle
sns.kdeplot(under60["drug_exposure_cycle"], label='Under 60', shade=True, ax=axs[0])
sns.kdeplot(over60["drug_exposure_cycle"], label='Over 60', shade=True, ax=axs[0])
axs[0].set_xlabel('Drug Exposure Cycle')
axs[0].set_ylabel('Density')
axs[0].set_title('Distribution of Drug Exposure Cycle by Age Group')
axs[0].legend()

# Average days supply
sns.kdeplot(under60["average_days_supply"], label='Under 60', shade=True, ax=axs[1])
sns.kdeplot(over60["average_days_supply"], label='Over 60', shade=True, ax=axs[1])
axs[1].set_xlabel('Average Days Supply')
axs[1].set_ylabel('Density')
axs[1].set_title('Distribution of Average Days Supply by Age Group')
axs[1].legend()

plt.tight_layout()
plt.show()
```
